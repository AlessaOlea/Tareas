```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import linalg as LA
```

```python
def get_data(N):
  x = np.linspace(0,2,N)
  y = np.zeros_like(x)
  eps = 1E-1
  for i in range(N):
    y[i] = 3*x[i] ** 2 + 5*x[i] - 2 + 5*(eps*(-1)**i)

  return x,y
```

```python
def plot_data(x, y, params):
  z = params[0] + params[1]*x + params[2]*x**2 #en la matriz de 2x1, params[0] representa al b (se desplaza hacia arriba arriba), y el params[1]*x representa a m (Abajo)
  plt.figure()
  plt.plot(x,y,"bh", label = "Samples") #Samples
  plt.plot(x,z,"-m", label = "Model") #Model
  plt.title("Cuadratic Regression") #Linear regression
  plt.xlabel("X Data",) #X data
  plt.ylabel("Y Data") #Y data
  plt.grid(lw=2)
  plt.legend()
  plt.show()

#plot_data(x, y, params)
```

```python
def Cuadratic_regression(N = 20):
  x,y = get_data(N)
  A = np.vstack([np.ones(len(x)), x, x**2]).T #Acomoda arreglos de tama√±o vertical y .T nos da la matriz transpuesta
  AtA = A.T@A #@ Multiplicador por default de matricez de python
  Aty = A.T@y
  params = LA.solve(AtA, Aty)
  plot_data(x,y,params)
  Ori_data = np.array([-2., 5., 3.])
  Error = LA.norm(params - Ori_data)
  print("Error: ", Error)
  return params
```

```python
Cuadratic_regression(N = 60)
```
