```python
#Analizar casos de ebola en un pais
Guinea =df[df["Country"] == "Guinea"] # Genara un nuevo dataframe solo con datos de guinea
Guinea.head()
Guinea.shape
Guinea.isnull().sum()
```

```python
Guinea.describe()
```

```python
Guinea.isnull().sum()
```

```python
#Llenar datos nulos con el promedio de la base de datos
mean_value = Guinea ["No. of confirmed deaths"].mean()
Guinea.fillna(mean_value, inplace = True)
Guinea.isnull().sum()
```

```python
#Grafica de casos confirmados (infectados) contra decesos
#Asignamos variables a los datos de las columnas de infectados y muertes
#Para visualizar el avance del ebola en guinea
#Se hizo uun slice para mostrar los datos del 10 en 10 (Grafica de barras)
Guinea_infectados = Guinea["No. of confirmed cases"][0:-1:10]
Guinea_muertes = Guinea["No. of confirmed deaths"][0:-1:10]
fechas = Guinea.index[0:-1:10]
plt.plot(fechas, Guinea_infectados, "o:r", ms = 4, label = "Infectados")
plt.plot(fechas, Guinea_muertes, "-b", ms = 4, label = "Muertos")
plt.title("Avance del Ebola en Guinea")
plt.xlabel("Fecha")
plt.ylabel("Número de casos")
plt.legend()
plt.grid(True)
plt.xticks(rotation = 70)
plt.savefig('ebola_guinea.png', dpi=300, format='png')
plt.show()
```

```python
#Analizar avance del ebola en niños, adultos y adultos mayores (ambos gneros)
Guinea_boys = Guinea["Boys"][0:-1:10]
Guinea_adults = Guinea["Adult Male"][0:-1:10]
Guinea_old = Guinea["Old Male"][0:-1:10]
fechas = Guinea.index[0:-1:10]

plt.plot(fechas, Guinea_boys, "o:r", ms = 4, label = "Niños")
plt.plot(fechas, Guinea_girls, "-b", ms = 4, label = "Adultos")
plt.plot(fechas, Guinea_adults, "-g", ms = 4, label = "Adultos mayores")

plt.title("Avance del Ebola en Hombres")
plt.xlabel("Fecha")
plt.ylabel("Número de casos")
plt.legend()
plt.grid(True)
plt.xticks(rotation = 70)
```
