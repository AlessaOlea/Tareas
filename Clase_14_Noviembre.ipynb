```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import linalg as LA
```

Ajuste de cuadrados minimos
tenemos una matriz
A=\begin{bmatrix}
1 & X_{0}  \\
1 & X_{1} \\
1 & x_{n} \\
\end{bmatrix}

Aᵀ*x \\
Matriz cuadrada e invertible X=[b,m], vector

```python
def get_data(N):
  x = np.linspace(0,2,N)
  y = np.zeros_like(x)
  eps = 1E-1
  for i in range(N):
    y[i] = 2*x[i] + 3 + (eps*(-1)**i)

  return x,y
```

```python
x,y = get_data(40)
def plot_data(x,y, params):
  z = params[0] + params[1]*x #en la matriz de 2x1, params[0] representa al b (se desplaza hacia arriba arriba), y eel params[1]*x representa a m (Abajo)

  plt.figure()
  plt.plot(x,y,"bh", label = "Samples") #Samples
  plt.plot(x,z,"-m", label = "Model") #Model
  plt.title("Linear Regression") #Linear regression
  plt.xlabel("X Data",) #X data
  plt.ylabel("Y Data") #Y data
  plt.grid(lw=2)
  plt.legend()
  plt.show()

#x,y = get_data(40)
#plot_data(x,y)
```

```python
def linear_regression(N = 20):
  x,y = get_data(N)
  A = np.vstack([np.ones(len(x)),x]).T #Acomoda arreglos de tamaño vertical y .T nos da la matriz transpuesta
  AtA = A.T@A #@ Multiplicador por default de matricez de python
  Aty = A.T@y
  params = LA.solve(AtA, Aty)
  plot_data(x,y,params)
  return params

#Lo que hacemos aqui es:
#Ax = y
#AtAx = Aty
#x = (AtA)^-1 * Aty
```
