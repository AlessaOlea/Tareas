```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 20)
y2 = x**2

plt.plot(x, y2, "h--b", mec = "g") #color marcador

fuente1 = {"family":"serif", "color":"black", "size":15} #definir una fuente
fuente2 = {"family":"serif", "color":"red", "size":15}

plt.title("Gráfica de x al cubo", fontdict = fuente1) #titulo de la gráfica
plt.yticks(fontsize = 10)
plt.xlabel("Eje x", fontdict = fuente2)
plt.xticks(fontsize = 10)
plt.ylabel("Eje y", fontdict = fuente2)

plt.grid() #ponerle cuadrilla
plt.xlim(-8, 8)
plt.ylim(0, 100)

plt.legend(["Función x al cuadrado"], loc = "upper center", fontsize = 10)

plt.show()

plt.savefig("Función x al cuadrado.png", dpi = 300, bbox_inches = "tight") #guardar imagen
```

```python
x = np.linspace(-10, 10, 200)

#1ra gráfica

y1 = np.sin(x**2)
plt.subplot(2,2,1)
plt.plot(x, y1, "r") # r = rojo

plt.grid(color = 'gray', linestyle = '--', linewidth = 1) #agrega cuadrilla

#2da gráfica

y2 = np.tanh(x)
plt.subplot(2,2,2)
plt.plot(x, y2, "b") # b = azul

plt.grid(color = 'gray', linestyle = '--', linewidth = 1) #agrega cuadrilla

#3ra gráfica

y3 = (np.sin(x))/x
plt.subplot(2,2,3)
plt.plot(x, y3, "g") # g = verde

plt.grid(color = 'gray', linestyle = '--', linewidth = 1) #agrega cuadrilla

#4ta gráfica

y4 = np.exp(-x/10) * np.sin(x)
plt.subplot(2,2,4)
plt.plot(x, y4, "y") # y = amarillo

plt.grid(color = 'gray', linestyle = '--', linewidth = 1) #agrega cuadrilla

plt.show() #imprimidor
```
